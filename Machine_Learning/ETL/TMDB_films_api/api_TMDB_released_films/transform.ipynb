```python
import numpy as np
import pandas as pd

import ast

import pyarrow.parquet as pq
```

# Read and merge dataframes

#### Read dataframe

```python
df_main = pd.read_csv('input/10k_main.csv', index_col=0)
df_main.head(5)
```

```python
df_main = df_main.rename(columns={'Movie_id':'id'})
df_main.head(3)
```

```python
df_cred = pd.read_parquet('input/10k_credits.parquet')
df_cred.head(5)
```

### Merge dataframes

```python
# enable to see all columns in a notebook
pd.set_option('display.max_columns', None)

# enable to see all rows in a notebook
pd.set_option('display.max_rows', None)
```

```python
df = pd.merge(df_main, df_cred, on='id')
df.head(5)
```

### Drop duplicates columns

```python
df = df.drop(columns=(['Revenue', 'Genres', 'budget']))
df.head(5)
```

# Create extra columns 

### Evaluate dtypes 

```python
# Function that safely evaluates a string representation of a literal Python expression and returns the corresponding Python object

list_items = ['Keywords', 'crew', 'cast-name', 'spoken_languages', 'genres', 'production_countries', 'production_companies']

for item in list_items:

    df[item] = df[item].apply(ast.literal_eval)
```

### Define functions

##### From list of dictionaries

```python
# Key words
def extract_from_list_dict(genre_list, key):
    result_list = [genre[key] for genre in genre_list]
    result = None if len(result_list) == 0 else result_list[0]
    return result


def add_col_list_dict(df,new_col_name, origin_col_name, key):
    df[new_col_name] = df[origin_col_name].apply(extract_from_list_dict, key=key)
```

##### From dictionaries

```python
# Professions
def add_col_dict(df, new_col_name, searched_item, searched_col_name):
    add_col = [key.get(searched_item, np.nan) for key in df[searched_col_name]]
    df[new_col_name] = add_col
```

##### From list - 2 items

```python
# Define a function to extract the first and second items from a list
def extract_2_items(actor_list):
    return actor_list[:2] if actor_list else [None, None]


def add_2_items_list(df, new_col_name1, new_col_name2, source_col_name):
    # Apply the function using a list comprehension to create two new columns
    df[[new_col_name1, new_col_name2]] = pd.DataFrame(df[source_col_name].apply(extract_2_items).tolist(), index=df.index)
```

##### From list - 1 item

```python
# Define a function to extract the first item from a list
def extract_1_item(actor_list):
    return actor_list[:1] if actor_list else [None]


def add_1_item_list(df, new_col_name1, source_col_name):
    # Apply the function using a list comprehension to create two new columns
    df[[new_col_name1]] = pd.DataFrame(df[source_col_name].apply(extract_1_item).tolist(), index=df.index)
```

### Call functions for extra cols

```python
# Keywords
add_col_list_dict(  df,
                    new_col_name='keyword', 
                    origin_col_name='Keywords', 
                    key='name'
                    )

df.head(1)
```

```python
# Professions
add_col_dict(df=df, 
        new_col_name='director', 
        searched_item='Director', 
        searched_col_name='crew'
        )

add_col_dict(df=df, 
        new_col_name='producer', 
        searched_item='Producer', 
        searched_col_name='crew'
        )

add_col_dict(df=df, 
        new_col_name='sound', 
        searched_item='Original Music Composer', 
        searched_col_name='crew'
        )

add_col_dict(df=df, 
        new_col_name='camera', 
        searched_item='Director of Photography', 
        searched_col_name='crew'
        )

add_col_dict(df=df, 
        new_col_name='screenplay', 
        searched_item='Screenplay', 
        searched_col_name='crew'
        )


df.head(2)
```

```python
# Actors
add_2_items_list(   df=df, 
                    new_col_name1='actor1', 
                    new_col_name2='actor2', 
                    source_col_name='cast-name'
                )

df.head(2)
```

```python
add_1_item_list(df=df, 
            new_col_name1='production_company', 
            source_col_name='production_companies'
            )


add_1_item_list(df=df, 
            new_col_name1='production_country', 
            source_col_name='production_countries'
            )


add_1_item_list(df=df, 
            new_col_name1='genre', 
            source_col_name='genres'
            )


add_1_item_list(df=df, 
            new_col_name1='spoken_language', 
            source_col_name='spoken_languages'
            )

df.head(2)
```

### Drop source columns

```python
df = df.drop(columns=(['Keywords', 'genres', 'crew', 'cast-name', 'spoken_languages', 'production_countries', 'production_companies']))
df.head(5)
```

#### Rename columns

```python
df = df.rename(columns={
                        'Budget' : 'budget'
                        })

df.head(3)
```
