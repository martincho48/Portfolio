```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Starting point

```python
df = pd.DataFrame({'col1': [ 1, np.nan, 3, 4, 5],

                    'col2':[np.nan, 20, 30, np.nan, 40],
                   
})


df.head(5)
```

### Add new col 'date' into existing df with date range 

```python
df['date'] = pd.date_range('2025-01-01', periods=5, freq='D')
df
```

### Add new col 'col3' using numpy range - use function where you define number of points

```python
df['col3'] = np.linspace(10, 20, 5)
df
```

### Show all NaN

```python
df[df.isna().any(axis=1)]
```

### Create new df with all NaNs dropped

```python
df2 = df.dropna().copy()
df2
```

### Create new df with all NaNs filled in

```python
df3 = df.fillna('pokus')
df3
```

### Fill all NaN in numeric cols with MEAN

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

### Fill  NaN in col2  with a value 1000

```python
df['col2'] = np.where(df['col2'].isna(), 'replaced', df['col2'])
df
```

### Plot line chart with markers and grid and size - define x and y axis, and define names of x and y axis

```python
df.plot(x='date', y='col1', figsize=(10,8), marker='o')

plt.xlabel('date')
plt.ylabel('y-value')
plt.grid(True)
plt.show()
```
