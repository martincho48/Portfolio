```python
import numpy as np
import pandas as pd
import random
```

```python
class Figure:

  def __init__(self, name, min_attack=5, max_attack=10, shield=2, life=200):
    self.name = name
    self.life = life
    self.dead = False
    self.min_attack = min_attack
    self.max_attack = max_attack
    self.shield = shield


  def hit(self):#attack function
    return random.randint(self.min_attack, self.max_attack) #randomly chooses power of attack

  def combo_hit(self):
    combo_attack=2*self.max_attack
    return combo_attack

  def super_combo_hit(self):
    super_combo_attack=2.5*self.max_attack
    return super_combo_attack

  def self_heal(self):
    combo_heal=1.5*self.life
    return combo_heal

  def super_self_heal(self):
    super_combo_heal=2.5*self.life
    return super_combo_heal

  def healing(self): #do nothing - it is here just not to be empty, it is for 'wizzard' character purposes
    healed_life=self.life
    return healed_life

#good_guy.life=0.8*bad_guy.life


  def block(self, hit_function):
    if hit_function > self.shield:
      self.life -= hit_function - self.shield #substracts from life if the hit is bigger than shield
      if self.life < 1:
        print(f"{self.name} is dead!")
        self.dead = True
```

```python
#parenting - create new character from the old ones and adding new power-'heal' and new super_combo

class wizzard(Figure):

  def __init__(self, name, min_attack=5, max_attack=10, shield=2, life=200, heal=30):

    self.heal=heal #new power
    
    super().__init__(name, min_attack, max_attack, shield, life) #link to parents - no need to define them again

  def combo_hit(self):
    combo_attack=2.2*self.max_attack
    return combo_attack

  def super_combo_hit(self): #self heal, double attack
    combo=2.7*self.max_attack
    return combo

  def healing(self):
    healed_life=self.life+(4*self.heal)
    return healed_life
```

```python
#The fight - define loop function

def mortal_combat(bad_guy, good_guy):
  while True:
    
    if good_guy.dead or bad_guy.dead:
      break

    draw=random.randint(0, 1)
  
    if draw==1:
     
      #COMBO attack - Good Guy
      if (bad_guy.life/good_guy.life)>=1.3 and (bad_guy.life/good_guy.life)<3:
        print(f"{good_guy.name} is attacking with COMBO-attack!")

        bad_guy.block(good_guy.combo_hit())
        
      #SUPER COMBO attack - Good Guy
      elif (bad_guy.life/good_guy.life)>=3:
        print(f"{good_guy.name} is attacking with SUPER COMBO!")
        
        bad_guy.block(good_guy.super_combo_hit())
        good_guy.life=good_guy.healing() # works just for wizzard

      #Normal attack - Good Guy
      else:
        print(f"{good_guy.name} is attacking!")
        bad_guy.block(good_guy.hit())
        

    if draw==0:
      #COMBO attack - Bad Guy
      if (good_guy.life/bad_guy.life)>=1.5 and (good_guy.life/bad_guy.life)<3:
        print(f"{bad_guy.name} is attacking-COMBO-self healing!")
        
        bad_guy.life=bad_guy.self_heal()

      #SUPER COMBO attack - Bad Guy
      elif (good_guy.life/bad_guy.life)>=1.5:
        print(f"{bad_guy.name} is attacking with SUPER COMBO!")

        bad_guy.life=bad_guy.super_self_heal()

      #Normal attack - Bad Guy
      else:
        print(f"{bad_guy.name} is attacking!")
        good_guy.block(bad_guy.hit())

    print(f"{good_guy.name}: {good_guy.life}")
    print(f"{bad_guy.name}: {bad_guy.life}\n")
```

```python
#define the figures:
good = Figure("Elf", 20, 40, 10, 300)
evil = Figure("Orc", 50, 70, 0, 300)
good_child = wizzard("Gandalf", 30, 50, 0, 300, 20)
```

```python
#The final fight

mortal_combat(evil, good_child)
```
